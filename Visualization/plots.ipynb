# Box Plot

```python
import numpy as np
import pandas as pd
pd.set_option('max_columns', 90)
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set_theme(style="whitegrid")
tips = sns.load_dataset("tips")
ax = sns.boxplot(x=tips["total_bill"])
```

```python
ax = sns.boxplot(x="day", y="total_bill", data=tips)
```

```python
ax = sns.boxplot(x="day", y="total_bill", hue="smoker",data=tips, palette="Set3")
```

```python
ax = sns.boxplot(x="day", y="total_bill", hue="time",data=tips, linewidth=2.5)
```

# Violin Plot


```python
sns.set_theme(style="whitegrid")
tips = sns.load_dataset("tips")
ax = sns.violinplot(x=tips["total_bill"])
```

```python
ax = sns.violinplot(x="day", y="total_bill", data=tips)
```

```python
ax = sns.violinplot(x="day", y="total_bill", hue="smoker",
                    data=tips, palette="muted")
```

```python
ax = sns.violinplot(x="day", y="total_bill", hue="smoker",
                    data=tips, palette="muted", split=True)
```

```python
ax = sns.violinplot(x="day", y="total_bill", hue="sex",
                    data=tips, palette="Set2", split=True,
                    scale="count")
```

# Strip Plot

```python
ax = sns.stripplot(x="total_bill",data=tips)
```

```python
ax = sns.stripplot(x="day", y="total_bill", data=tips)
```

```python
ax = sns.stripplot(x="sex", y="total_bill", hue="day", data=tips)
```

```python
ax = sns.stripplot(x="day", y="total_bill",hue="sex", data=tips)
```

```python
ax = sns.stripplot(x="day", y="total_bill",hue="sex", dodge=True,data=tips)
```
